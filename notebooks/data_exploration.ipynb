```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the processed data
train_data = pd.read_json('../data/processed/train_data.json')
val_data = pd.read_json('../data/processed/val_data.json')
```

```python
# Display basic information about the datasets
print("Training data shape:", train_data.shape)
print("Validation data shape:", val_data.shape)
```

```python
# Display sample data
print(train_data.head())
```

```python
# Analyze question types
plt.figure(figsize=(10, 6))
train_data['qtype'].value_counts().plot(kind='bar')
plt.title('Distribution of Question Types')
plt.xlabel('Question Type')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze text length
train_data['text_length'] = train_data['text'].str.len()
plt.figure(figsize=(10, 6))
sns.histplot(data=train_data, x='text_length', bins=30)
plt.title('Distribution of Text Length')
plt.xlabel('Text Length')
plt.show()
```
